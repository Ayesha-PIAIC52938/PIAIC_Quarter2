## <div class= "alert alert-info">1. apply()
    
Pandas Series<span style="color:red"> .apply() </span> function invoke the passed function on each element of the given series object.    

Using <span style="color:red"> .apply() </span>method we can execute a function to a single column (a series), all (dataframe). 
    
    

```python
def div(a):
    return a/2
```

```python
sa = new.apply(div)  # calling div function to divided each value of new by 2 and stored it in sa
sa
```

each value of the series is passed to the div function that divides it by 2 and return it to .applay() that replaces the orignal value by the divided value

We create functions that are reusable in the code. thus for .apply() instead of defining a div function we can use a lambda function 

```python
sa_lambda = new.apply(lambda series_value: series_value/2 )  # same code via lambda function
sa_lambda
```

we can apply logic in lambda functions too:

```python
sa_lambda1 = new.apply(lambda series_value:  np.where(series_value > 60, "Pass", "Fail") )  # same code via lambda function
sa_lambda1
```

.apply() applies whatever the function is on the series values, thus x in the above code represents the values in the series

## <div class= "alert alert-info">2. sort_values()
    
<span style="color:red"> .sort_values()  </span> function is used to sort values on Series object. It sorts the series in ascending order or descending order, by default it does in ascending order. You can specify your preference using the ascending parameter which is True by default.

<b> Series.sort_values(axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None) </b>
    


```python
s1 = pd.Series(['JAVA','PySpark','Spark','Hadoop'])     # creating a series
s2 = pd.Series([1,3,2,2,5,np.nan])  
print(s1,'\n')
print(s2)
```

#### Sort pandas Series in an ascending order:

```python
sortedseries1 = s1.sort_values()      # <or>   ortedseries = series.sort_values(ascending=True)
sortedseries1
```

```python
sortedseries2 = s2.sort_values(ascending=True)
sortedseries2
```

#### Sort pandas Series in an descending order:

```python
sortedseries1 = s1.sort_values(ascending=False)
sortedseries1
```

```python
sortedseries2 = s2.sort_values(ascending=False)
sortedseries2
```

<mark> .sort()  places NaN values at last by default to place it at top, set na_position=’first’ to sort by placing all NaN values at the first.:</mark>

```python
sortedSeries2 = s2.sort_values(ascending=True, na_position='first')
sortedSeries2
```

#### Sort pandas Series inplace:

Note that the sort_values() function returns a sorted copy of the original series. To modify the original series in-place, pass inplace=True to the function.

```python
s1.sort_values(ascending=False, inplace=True)     # sorts s1 (desending order) and save the sorted series in s1
s1
```

```python
s2.sort_values( inplace=True)     # sorts s2 (ascending order) and save the sorted series in s1
s2
```

## <div class= "alert alert-info">3. sort_index()


pandas <span style= "color:red">.sort_index()</span> function is used to sort the pandas series/DataFrame by index or columns by name/labels. This function takes several parameters like axis, level, ascending, inplace, kind, na_position, sort_remaining, ignore_index, and key and returns a new DataFrame with the sorted result. Use inplace=True to update the existing DataFrame.

<b>sort_index() key Points</b>

- Applied soring on axis, not on data.
- Use axis=1 to sort by column names, axis-0 to sort by index.
- Supports different sorting algorithms ‘quicksort’, ‘mergesort’, ‘heapsort’, ‘stable’

<b> .sort_index(axis=0, level=None, ascending=True, inplace=False, kind='quicksort', na_position='last', sort_remaining=True, ignore_index=False, key=None)</b>


```python
s1
```

## <div class= "alert alert-info">16  iloc()

<span style = "color:red">.iloc[]</span> is a property that is used to select series values or rows and columns in case of data frame by position/index. If the position/index does not exist, it gives an index error. 

pandas loc[] is another property that is used to operate on series value or the column and row labels of dataframe.
    
    
    
    For a better understanding of these two learn the differences and similarities between pandas loc[] vs iloc[]. The difference between loc[] vs iloc[] is described by how you select rows and columns from pandas DataFrame.
    
    
Pandas DataFrame is a two-dimensional tabular data structure with labeled axes. i.e. columns and rows. Selecting columns from DataFrame results in a new DataFrame containing only specified selected columns from the original DataFrame.
