### - <ins>Replacing rows with missing data</ins>: 

Two ways to replace missing values:
- Insert some value instead of missing value.The <span style="color:red">
fillna( )
    </span> method allows us to replace missing value/empty cell/NaN/Null with some value:
- Replace missing value with the mode, median or mean of the column from which its missing

<b> Syntax:</b>

##### Series.<span style="color:red">fillna(</span><span style="color:green"> self, value=None, method=None, axis=None, inplace=False, limit=None, downcast=None, **kwargs</span><span style="color:red">)</span> 


- 'backfill' or 'bfill':   fill NaN with previous not NaN value
- 'ffill':   fill NaN with next not NaN value
- 'pad':   fill NaN with last not NaN value (use when there are two or more consective NaN values)

#####  <span style="color:green"> Replacing missing value by mean, median or mod: </span> 
 
 Pandas uses the <span style="color:red">
 mean(), median() and mode()
    </span> 
    
 One of the key points is to decide which technique out of the above-mentioned to use to get the most effective value for the missing values. The central tendency measure such as mean, median, or mode is considered for imputation. The goal is to find out which is a better measure of the central tendency of data and use that value for replacing missing values appropriately.  

<span style='background :yellow' >Pros:

- Prevent data loss which results in deletion of rows or columns
- Works well with a small dataset and is easy to implement.

<span style='background :yellow' >Cons:

- Works only with numerical continuous variables.
- Can cause data leakage
- Do not factor the covariance between features.    
    
#### when to use mean(), median() or mode()?
    
Real-world distributions are usually <b> skewed </b>, if there’s too much skewness in the data, then many statistical models don’t work effectively because in skewed data, the tail region may act as an <b> outlier </b> for the statistical model, and we know that outliers adversely affect a model’s performance, especially regression-based models. While there are statistical models that are robust enough to handle outliers like tree-based models, you’ll be limited in what other models you can try. 
    
<b> Skewed data</b> is data that creates an asymmetrical, skewed curve on a graph. In statistics, the graph of a data set with normal distribution is symmetrical and shaped like a bell. However, skewed data has a "tail" on either side of the graph
    
<b> outliers </b> datapoints that make the overall data skewed. Outliers data points have a significant impact on the mean and hence, in such cases, it is not recommended to use the mean for replacing the missing values. Using mean values for replacing missing values may not create a great model and hence gets ruled out. 

- <b> mean():</b> For symmetric data distribution, one can use the mean value for replacing missing values.   
    
- <b>median():</b> For asymmetric data distribution (skewed data), one can use the mean value for replacing missing values. 
    
- <b> mod():</b> For asymmetric data distribution (skewed data), one can use the mean value for replacing missing values. 
    
#####  <ins><span style="color:blue"> Replacing by mean<ins>:</span> 
One of the techniques is to replace the missing values with the mean value of the entire feature column. In the case of skewed data, it may not be a good idea to use mean() for replacing the missing values. Note that replacing missing data with mean values can only be done with numerical data.
    
##### <ins><span style="color:blue"> Replacing by median<ins>:</span>
    
Another technique is replacing the missing values with the median() value of the entire feature column. When the data is skewed, it is good to consider using the median value for replacing the missing values. Note that replacing missing data with median value can only be done with numerical data.  
    
#####  <ins><span style="color:blue"> Replacing by mode<ins>: <span> 
    
Yet another technique is mode replacing the missing values with the mode() value or most frequent value of the entire feature column. When the data is skewed, it is good to consider using mode values for replacing the missing values. For data points such as the salary field, you may consider using mode for replacing the values. Note that replacing missing data with mode values can be done with numerical and categorical data.         

```python
data.Year.isnull().sum()   # year column has 26 NaN values
```

```python
data.Year.mode()
```

```python
data.Year = data.Year.fillna(data.Year.mode()[0]) # without [0] Year NaN values are not replaced with mode= 1978.0
```

```python
data.Year.isnull().sum()    # NaN:True =1 and not NaN: False = 0
```

## To fill multiple series with one go use dictionaries with fillna()

```python
data = data.fillna({
            'Group':data.Group.mode()[0],
            'alma_mater':data.alma_mater.mode()[0],
            'undergraduate_major':data.undergraduate_major.mode()[0],
            'graduate_major':data.graduate_major.mode()[0],
            })
```

###  - <ins>Removing rows with missing data</ins>:

To remove row with missing values we use <span style="color:red">
dropna()
    </span> method. 
- dropna() method returns a new DataFrame, and will not change the original.
- dropna( inplace = True) method remove missing value from the orignal dataset and return nothing



<span style="color:red">dropna(</span><span style="color:green">axis: 'Axis' = 0,
    how: 'str' = 'any',
    thresh=None,
    subset: 'IndexLabel' = None,
    inplace: 'bool' = False</span>
<span style="color:red">)</span>
Docstring:
Remove missing values.

See the :ref:`User Guide <missing_data>` for more on which values are
considered missing, and how to work with missing data.

Parameters
----------
<b>axis :</b> {0 or 'index', 1 or 'columns'}, default 0
    Determine if rows or columns which contain missing values are
    removed.

- 0, or 'index' : Drop rows which contain missing values.
- 1, or 'columns' : Drop columns which contain missing value.

- Pass tuple or list to drop on multiple axes. Only a single axis is allowed.

<b>how :</b> {'any', 'all'}, default 'any'.Determine if row or column is removed from DataFrame, when we have
    at least one NA or all NA.

- 'any' : If any NA values are present, drop that row or column.
-'all' : If all values are NA, drop that row or column.

<b>thresh : int, optional. Require that many non-NA values.

<b>subset :</b> column label or sequence of labels, optional. Labels along other axis to consider, e.g. if you are dropping rows
    these would be a list of columns to include.
    
<b>inplace :</b> bool, default False.  If True, do operation inplace and return None.

Returns
-------
DataFrame or None
DataFrame with NA entries dropped from it or None if ``inplace=True``.




<b><span style='background :yellow' >Pros:

- A model trained with the removal of all missing values creates a robust model.
    
<b><span style='background :yellow' >Cons:

- Loss of a lot of information.
- Works poorly if the percentage of missing values is excessive in comparison to the complete dataset.

### <div class= "alert alert-warning">(3) How to Get Unique Values from a DataFrame Column?<div>

Consider a scenario where you have a dataframe consisting of students with respect to the subjects they study. In such a condition, if you wish to identify the total number of subjects being studied overall, it is quite infeasible for you to count each row of the dataframe and identify the unique subject category. To reduce your manual work, below are the two methods by which you can easily get unique values in a column of pandas dataframe:

- Using unique() method     
- Using value_counts() method

#### <div class = "alert alert-danger">unique() method :</div>
    
unique() method is used when we deal with a single column of a DataFrame (i.e a series) and returns all unique elements of a column. The method returns a DataFrame containing the unique elements of a column, along with their corresponding index labels.
Uniques are returned in order of appearance. unique() also take NaN as a class.

```python
data.Gender.unique() # shows the value that could be assigned to each usertype i.e Male or Female

#out of the 357 values, usertypr could be assign Male or Female
```

```python
import numpy as np
df = pd.DataFrame({"A":["Sandy", "alex", "brook", "kelly", np.nan],
                   "B":[np.nan, "olivia", "olivia", " ", "amanda"], 
                   "C":[20 + 5j, 20 + 5j, 7, None, 8],
                   "D":[14.8, 3, None, 6, 6]})      

df
```

```python
df.B.unique()
```

```python
df.B.nunique()
```

#### <span style = "color:green"> In ML we call Male and Female as classes, so basically unique() tell us about classes in a series

### <mark> Difference between nunique() and unique()
<mark>The difference is: The unique() function returns all the unique values of the column in the form of an array .
The nunique() function returns the number of all unique values . Also nunique() dont count NaN values

#### <div class = "alert alert-danger"> value_counts() method:</div>
 
The value_counts() method returns a Series containing the counts of unique values. This means, for any column in a dataframe, this method returns the count of unique entries in that column. Simply put value_counts() method returns occurences of unique value in a DataFrame column (series)

```python
data.Gender.value_counts()
```
