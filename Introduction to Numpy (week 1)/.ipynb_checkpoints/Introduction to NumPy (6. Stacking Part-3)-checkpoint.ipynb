# <div class="alert alert-success" >(6). Stacking (part-3)

```python
import numpy as np
```

#### <div class= "alert alert-info"> Horizontal stacking
    
<span style="color:red">hstack()</span> function is basically <span style="color:red">concatenate()</span> with axis = 1. horizontal stacking results in a new array with same rows as array1/array2 but columincreases i.e new array extends horizontally
    


#### <div class= "alert alert-danger">1D horizontal stacking

```python
a = np.arange(1,6)
b = a*2
print("a =",a,'\n')
print("b =",b)
```

```python
arr1d = np.hstack((a,b)) 
arr1d
```

<mark> resultant 1D array stacks b after a ie [a b] </mark>



#### <div class= "alert alert-danger">2D horizontal stacking


```python
c = np.arange(10).reshape(5,2)
d = c*3
print("c =",c,'\n')
print("d =",d)
```

```python
arr2d = np.hstack((c,d)) 
arr2d
```

<mark> resultant 2D array stacks b after a ie [c^T d^T]^T </mark>



#### <div class= "alert alert-danger">3D horizontal stacking


```python
e = np.arange(20).reshape(2,2,5)
f = e*3
print("e =",e,'\n')
print("f =",f)
```

```python
arr3d = np.hstack((e,f)) 
arr3d
```

<mark> resultant 3D array stacks corrosponding rows of f after e per depth ie di = [[ e[di], f[di] ]]^T </mark>



#### <div class = "alert alert-warning"> Column stacking
    
<span style="color:red"> Column_stack()</span> stacks array2 as extended columns of array1. column_stack() is just like hstack().   

#### <div class= "alert alert-danger">1D column stacking

```python
a1 = np.arange(1,6)
b1 = a1*2
print("a1 =",a1,'\n')
print("b1 =",b1)
```

```python
arr1d_cs = np.column_stack((a1,b1))  
```

<mark> resultant 1D array stacks b1 after a1 ie [a1 b1]^T </mark>



#### <div class= "alert alert-danger">2D column stacking

```python
c1 = np.arange(6).reshape(2,3)
d1 = c1*3
print("c1 =",c1,'\n')
print("d1 =",d1)
```

```python
arr2d_cs = np.column_stack((c1,d1)) 
arr2d_cs
```

<mark> resultant 2D array stacks d1 after c1 ie [ [c1 d1] ]</mark>



#### <div class= "alert alert-danger">3D column stacking

```python
e1 = np.arange(20).reshape(2,2,5)
f1 = e1*2
print("e1 =",e1,'\n')
print("f1 =",f1)
```

```python
arr3d_cs = np.column_stack((e1,f1))  
arr3d_cs
```

<mark> resultant 3D array stacks di of f1 after di of e1 ie di = [ [ dei dfi ] ]</mark>



#### <div class="alert alert-success" > summary

1D hstack()

- 1D array: array =[array1, array2]     

2D hstack()

- resultant 2D array stacks b after a ie [c^T d^T]^T    
    
3D hstack()

- 3D array with depth: di = [[ a[di], b[di] ]]^T
