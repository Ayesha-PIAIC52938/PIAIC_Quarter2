#  <div class="alert alert-success" >Additional Numpy methods:

```python
import numpy as np
```

### <div class="alert alert-warning" >(1) np.repeat()
    
    

```python
arr = np.array([1,2,3,4,5,6])
arr
```

```python
np.repeat(arr,3)  # repeats each array element 3 times
```

```python
a =  np.array([[1,2],[3,4]])
a
```

```python
np.repeat(a,3,axis = 0)  # repeats each row 3 times
```

```python
np.repeat(a,3,axis = 1)  # repeats each colum 3 times
```

```python
np.repeat(a, [1, 2], axis=0) # first row repeated once and 2nd row twice
```

```python
np.repeat(a, [2, 1], axis=0) 
```

```python
np.repeat(a, [1, 2], axis=1) # first column repeated once and 2nd column twice
```

### <div class="alert alert-warning" >(2) np.tile()
    
    

```python
a = np.array([0, 1, 2]) 
a
```

```python
np.tile(a, 2)   # repeate the whole array 2 times
```

```python
np.tile(a, (2, 2))  #  makes a 2D array with each row as a repeate of the whole a array 2 times
```

```python
np.tile(a, (2, 1, 2))  #  makes a 3D array with each page as a repeate of the whole a array 2 times
```

```python
b = np.array([[1, 2], [3, 4]])
b
```

```python
np.tile(b, 2)    # np.tile(b, 2) = np.hstack(b,b)
```

```python
np.tile(b, (2, 1))   # np.tile(b, (2,1)) = np.vstack(b,b)
```

```python
np.tile(b, (2, 2))    
```

### <div class="alert alert-warning" >(3) np.all()


    
<b>Not a Number (NaN), positive infinity and negative infinity evaluate to True because these are not equal to zero.    

```python
np.all([[True,False],[True,True]])
```

```python
np.all([[True,False],[True,True]], axis=0)
```

```python
np.all([-1, 4, 5])
```

```python
np.all([1.0, np.nan])
```

```python
np.all([[True, True], [False, True]], where=[[True], [False]])
```

### <div class="alert alert-warning" >(4) np.unique()
    



    

```python
np.unique([1, 1, 2, 2, 3, 3])   # unique elements
```

```python
a = np.array([[1, 1], [2, 3]])

np.unique(a)      # unique elements
```

```python
a = np.array([[1, 0, 0], [1, 0, 0], [2, 3, 4]]) 

np.unique(a, axis=0)   # unique row
```

### <div class="alert alert-warning" >(5) np.sort()
    


    

```python
a = np.array([[1,4],[3,1]])
np.sort(a)                # sort along the last axis
```

```python
np.sort(a, axis=None)     # sort the flattened array
```

```python
np.sort(a, axis=0)        # sort along the first axis
```
