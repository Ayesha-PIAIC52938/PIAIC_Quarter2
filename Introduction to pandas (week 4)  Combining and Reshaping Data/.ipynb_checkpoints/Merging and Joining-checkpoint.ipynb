# <div class="alert alert-success" > Merging and Joining

```python
import pandas as pd
import numpy as np
import datetime as dt
```

To understand merge we'll first create two dataframes: customers and orders

```python
customers  = {'customer_id':[1234,1235,1111,2222,3334,3333,4444,5555],
             'Name':['Mike', 'Maria', 'Nasir', 'Ali',"Ahmed", 'Saad',"Hamid", "Kami"],
             'Address':['abc23-1', 'xyz-3','abc23-2', 'xyz-5','abc23-6', 'xyz-0','abc23-11', 'xyz-301'],
             }

customers = pd.DataFrame(customers)
customers
```

```python
orders = {'customer_id':[1234,1235,1234,1234,1235,1267,1237,1890],
          'order_id':[11,22,33,44,55,66,77,88],
         'order_date':[dt.date(2023,2,1),dt.date(2023,1,1),
                       dt.date(2023,2,2),dt.date(2023,1,1),
                       dt.date(2023,2,3),dt.date(2023,1,2),
                       dt.date(2023,2,4),dt.date(2023,1,2)],
         }

orders = pd.DataFrame(orders)
orders
```

### <mark>Why not use concat() to concatenate two dataframes? why use .merge()?

.merge() because it doesnt require two dataframes to have the same shape to be joined, whereas .concat() will give erroe if both dataframes shape not same

- we can roughly say that .merge() joins dataframes key wise where as concat() joins dataframes row/column wise

### <div class= "alert alert-warning">  .merge() methods: 

<b> Primary Key:</b> In database primary key is a uniwue identifier that uniquely identifies a row.

<b> Forigen Key:</b> primary key of one database repeated as a column in another database
    
There're are different types of merge and its important to know when to use which merge. Different merge types are:
  
  - Inner merge (default merge set to inner merge)[inner>>intersection i.e overlap]  
  - Outer merge  
  - Left merge
  - Right merge    
    
- For merge there should be at least one column that is present in both dataframes
- Default merg if outer merge i.e merge search for overlapping data in the same columns

#### Merging stepe for <b>.merge( orders , customers )</b>:

    
1. searches for any column of customer (right) dataframe that is also in the orders (left) dataframe (i.e customer_id column)
2. Make the same column (customer_id) as the first column with datapoints that are intersection of customer_id colum's datapoints of both customers and orders dataframes. customer_id is followed by columns of orders dataframe.
3. After orders dataframe columns, customers colums are placed with datapoints corrosponding to the overlapping datapoints of both frames customer_id column. 


### <div class= "alert alert-warning">Inner merge  (inner>> intersection)

```python
merged_data_inner = pd.merge(orders,customers, how = 'inner')  # by default inner join so writing  how = 'inner' is optional
merged_data_inner
```

```python
pd.merge(customers,orders)    # reversing parameter orders
```

#### By default if dont mention a column to merge on it merges on colum with same header in both dataframes. in the above example customer_id was the same header in both dataframes



### <div class= "alert alert-warning">  Outer merge (outer >> Union)

```python
merged_data_outer = pd.merge(customers,orders, how = 'outer')  #  how = 'outer' : union of data points in the same column
merged_data_outer
```

outer join/ merge works on the princile of alignment we studied in the series

### <div class= "alert alert-warning">  left merge

Left merge use only keys from left dataframe, similar to a SQL left outer join; preserve key order.    
    
    
In left merge priority is given to the left dataframe i.e orders. By priority we means that if only datapoints of left dataframe same columns are merged (orders columns first followed by customers columns) those values in common column against which right dataframe has no datapoints, NaN is written    

```python
customers
```

```python
orders
```

```python
merged_data_left = pd.merge(customers,orders, how = 'left') 
merged_data_left
```

<b> note that in left merge orders (left dataframe) is placed as it is followed by customers dataframe relavent data

how = 'left' : customer_id column overlappiing datapoints plus remaning datapoints from customers is picked and NaN values given to missing data of right dataframe i.e orders 



### <div class= "alert alert-warning"> right merge
    
Right merge use only keys from right dataframe, similar to a SQL right outer join;preserve key order.    

```python
customers
```

```python
orders
```

```python
merged_data_right= pd.merge(customers,orders, how = 'right')  
merged_data_right
```

how = 'right' : customer_id column's overlapping datapoints plus order dataframe customer_id is picked as it is and NaN placed for missing datapoints in left dataframe i.e customers

### <div class= "alert alert-warning">what if no common column in dataframes for merge? or column datapoints are same but column names are different
    


```python
customers  = {'customerID':[1234,1235,1111,2222,3334,3333,4444,5555],
             'Name':['Mike', 'Maria', 'Nasir', 'Ali',"Ahmed", 'Saad',"Hamid", "Kami"],
             'Address':['abc23-1', 'xyz-3','abc23-2', 'xyz-5','abc23-6', 'xyz-0','abc23-11', 'xyz-301'],
             }

customers = pd.DataFrame(customers)

orders = {'customer_id':[1234,1235,1234,1234,1235,1267,1237,1890],
          'order_id':[11,22,33,44,55,66,77,88],
         'order_date':[dt.date(2023,2,1),dt.date(2023,1,1),
                       dt.date(2023,2,2),dt.date(2023,1,1),
                       dt.date(2023,2,3),dt.date(2023,1,2),
                       dt.date(2023,2,4),dt.date(2023,1,2)],
         }

orders = pd.DataFrame(orders)
```

```python
customers
```

```python
#  pd.merge(customers,orders)        # will give merge error 
```

####  MergeError: No common columns to perform merge on. Merge options: left_on=None, right_on=None, left_index=False, right_index=False


in this case we'll specify columns of left and right dataframe for merge:

```python
merged = pd.merge(customers,orders,left_on='customerID', right_on='customer_id') # when column names are different but same data
merged
```

<b>Note that noe customerID and customer_id column are written twice (inner merge by default)

This case rarely happens. If by any chance it does happen we rename the columns. Above method can also be used if you purposefully dont want to merge on common column

### <div class= "alert alert-warning">what if more then one common column found dataframes for merge? 

```python
customers  = {'customerID':[1234,1235,1111,2222,3334,3333,4444,5555],
             'Name':['Mike', 'Maria', 'Nasir', 'Ali',"Ahmed", 'Saad',"Hamid", "Kami"],
             'Address':['abc23-1', 'xyz-3','abc23-2', 'xyz-5','abc23-6', 'xyz-0','abc23-11', 'xyz-301'],
              'order_date':[dt.date(2023,2,1),dt.date(2023,1,1),
                       dt.date(2023,2,2),dt.date(2023,1,1),
                       dt.date(2023,2,3),dt.date(2023,1,2),
                       dt.date(2023,2,4),dt.date(2023,1,2)]
             }

customers = pd.DataFrame(customers)

orders = {'order_id':[11,22,33,44,55,66,77,88],
          'customerID':[1234,1235,1234,1234,1235,1267,1237,1890],
         'order_date':[dt.date(2023,2,1),dt.date(2023,1,1),
                       dt.date(2023,2,2),dt.date(2023,1,1),
                       dt.date(2023,2,3),dt.date(2023,1,2),
                       dt.date(2023,2,4),dt.date(2023,1,2)],
         }

orders = pd.DataFrame(orders)
```

```python
customers
```

```python
orders
```

```python
pd.merge(customers,orders)
```

It merges on the first common column it encounters in both dataframes i.e customerID.

For example merging customers to order:

```python
pd.merge(orders,customers)    # Now order_id is the common column as its searched first in customers dataframe
```

### To merge on <span style="color:blue"> order_date</span></b> regardless of dataframes order

```python
pd.merge(customers,orders, on='order_date')
```

<b> Note that customerID was the other common colum with same name too. so suffex _x for left dataframe (customers) and _y for right dataframe (orders) are added to it. this is done because dataframe cant have two columns with same name.
    
to change the suffex _x ,_y to something that make more sense, do the following:    

```python
pd.merge(customers,orders, on='order_date',suffixes=("_customers", "_orders"))
```

#### To merge on multiple common columns: 

```python
pd.merge(customers,orders, on=['order_date', 'customerID'],suffixes=("_left", "_right"))
```

Now for merging it will first go to order_date columns, then the customerID column

## <div class= "alert alert-danger">.pivot()
    
Reshape data (produce a "pivot" table) based on column values. Uses unique values from specified `index` / `columns` to form axes of the resulting DataFrame. This function does not support data aggregation, multiple values will result in a MultiIndex in the columns. 
    
<b>Parameters

- <b>index</b> : str or object or a list of str, optional Column to use to make new frame's index. If None, uses existing index. Also accept list of index names.

- <b>columns</b> : str or object or a list of str Column to use to make new frame's columns. Also accept list of columns names.

- <b>values</b> : str, object or a list of the previous, optional Column(s) to use for populating new frame's values. If not specified, all remaining columns will be used and the result will have hierarchically indexed columns.

<b>Returns: Returns reshaped DataFrame.

<b> Raises</b> ValueError: When there are any `index`, `columns` combinations with multiple values. `DataFrame.pivot_table` when you need to aggregate.


<b>Notes:</b> For finer-tuned control, see hierarchical indexing documentation along with the related stack/unstack methods.    
    
### Syntax:
    f.pivot(index=None, columns=None, values=None) 

```python
d = {'foo': ['one', 'one', 'one', 'two', 'two','two'],
     'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
     'baz': [1, 2, 3, 4, 5, 6],
     'zoo': ['x', 'y', 'z', 'q', 'w', 't']}

df = pd.DataFrame(d)
df
```

```python
df.T # just rotates the dataframe 
```

pivot asks for which column to take as index, which to take as column names and which to take as values of those columns:

```python
df.pivot(index='foo', columns='bar', values='baz')
```

```python
df.pivot(index='foo', columns='bar', values=['baz', 'zoo'])  # list of columns for values
```

```python
df.pivot(index='foo', columns='bar')  #  values not specified takes the remaning columns as values
```

```python
df.pivot(index='foo', columns='zoo') 
```
